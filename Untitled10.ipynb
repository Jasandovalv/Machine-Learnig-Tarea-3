```python
from google.colab import files
import pandas as pd

# Subir archivo desde tu PC
uploaded = files.upload()

# Cargar datos desde el archivo subido
df = pd.read_excel(list(uploaded.keys())[0])

print("Archivo cargado:", list(uploaded.keys())[0],sep=';')
print("Dimensiones del dataset:", df.shape)

# Información general del dataset
df.info()

```

```python
df_seleccionado = df[['CR_DMA','Mes','Buque']]
print(df_seleccionado)
```

```python
df_seleccionado_rellenado = df_seleccionado.fillna(0)
print("DataFrame con NaNs reemplazados por cero:")
print(df_seleccionado_rellenado)
```

```python
df_seleccionado_rellenado.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'Mes' and sum 'CR_DMA'
cr_dma_por_mes = df_seleccionado_rellenado.groupby('Mes')['CR_DMA'].sum().reset_index()

# Define the order of months
month_order = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Convert 'Mes' to a categorical type with the specified order
cr_dma_por_mes['Mes'] = pd.Categorical(cr_dma_por_mes['Mes'], categories=month_order, ordered=True)

# Sort the DataFrame by the ordered 'Mes' column
cr_dma_por_mes = cr_dma_por_mes.sort_values('Mes')

plt.figure(figsize=(12, 6))
sns.barplot(x='Mes', y='CR_DMA', data=cr_dma_por_mes, palette='viridis')
plt.title('CR_DMA por Mes')
plt.xlabel('Mes')
plt.ylabel('Suma de CR_DMA')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filtrar el DataFrame para mostrar solo los buques con CR_DMA > 0
buques_con_cr_dma = df_seleccionado_rellenado[df_seleccionado_rellenado['CR_DMA'] > 0]

# Mostrar los buques y sus CR_DMA correspondientes
print("Buques que coincidieron con CR_DMA (valor > 0):")
print(buques_con_cr_dma[['Buque', 'Mes', 'CR_DMA']].drop_duplicates())
```
